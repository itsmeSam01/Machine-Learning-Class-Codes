```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
dataset_url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
data = pd.read_csv(dataset_url)

print("Dataset Preview (Entropy):")
print(data.head())

features = data.drop(columns=['species'])  
target = data['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

dt_model_entropy = DecisionTreeClassifier(criterion='entropy', random_state=42)
dt_model_entropy.fit(X_train, y_train)

y_pred_entropy = dt_model_entropy.predict(X_test)

accuracy_entropy = accuracy_score(y_test, y_pred_entropy)
print("\n=== Decision Tree (Entropy) ===")
print(f"Accuracy: {accuracy_entropy:.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_entropy))
```

```python
plt.figure(figsize=(20, 10))
tree.plot_tree(dt_model_entropy, feature_names=features.columns, class_names=dt_model_entropy.classes_, filled=True)
plt.title("Decision Tree (Entropy)")
plt.show()
```
